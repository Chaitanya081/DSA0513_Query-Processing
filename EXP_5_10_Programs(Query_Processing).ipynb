EXP_4 Write a Pandas program to create a line plot of the historical stock prices of Alphabet Inc. between two specific dates.

```python
import pandas as pd
import yfinance as yf
import matplotlib.pyplot as plt
```

```python
df=yf.download('GOOGL',start='2020-01-01',end='2020-12-31')
print(df.head())
df['Close'].plot(title="Stock Prices of Alphabet Inc",figsize=(10,5))
plt.xlabel("Date")
plt.ylabel("Price")
plt.show()
```

EXP_6 To create a scatter plot of the trading
volume/stock prices of Alphabet Inc. stock between two specific dates.

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv('data4.csv')
print(df.head())
```

```python
df['Date']=pd.to_datetime(df['Date'],format='%d-%m-%Y')
data=df[(df['Date']>='2020-04-01')& (df['Date']<='2020-04-30')]
plt.scatter(data['Date'],data['Close'])
plt.title("Alphabet Stock Price Vs Trading Volume")
plt.xlabel('Volume')
plt.ylabel('Close Price')
plt.show()
```

EXP_7 To create a Pivot table and find the maximum
and minimum sale value of the items

```python
import pandas as pd
import matplotlib as plt
```

```python
df=pd.read_csv('data5.csv')
print(df.head())
```

```python
pivot=pd.pivot_table(df,index='Item',values='Sale_amt',aggfunc=['min','max'])
print(pivot)
```

EXP_10 Create a dataframe of ten rows, four columns with random values. Write a Pandas program to highlight the negative numbers red and positive numbers black.

```python
import pandas as pd
import numpy as np
```

```python
df=pd.DataFrame(np.random.randn(10,4),columns=['A','B','C','D'])
print(df)
```

```python
def highlight_negative(val):
  if val>0:
    return 'color:black'
  else:
    return 'color:red'
styled_df=df.applymap(highlight_negative)
print(df)
styled_df
```
