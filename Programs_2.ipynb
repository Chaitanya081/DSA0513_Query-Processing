EXP_2 Write a Pandas program to display the IDs of employees who have
done two or more jobs in the past.

```python
import pandas as pd
```

```python
df=pd.read_csv("data2.csv")
print(df.head())
result=df.groupby("EMPLOYEE_ID").size()
result1=result[result>=2]
print("The Employees are Found:\n")
print(result1)
```

EXP_3 Write a Pandas program to display the details of jobs in descending
sequence on job title.

```python
import pandas as pd
```

```python
df=pd.read_csv("data3.csv")
print(df.head())
```

```python
result=df.sort_values(by="JOB_TITLE",ascending=False)
print(result)
```
